```python
from json import load
from pathlib import PurePath
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as s
import wordcloud as wordcloud
from PIL import Image
```

```python
user_json_path = PurePath("../../data/processed_c2_data/")
user_raw_path = PurePath("../../data/clarkson2_files/")
user_list = [user for user in os.listdir(user_raw_path) if user != ".DS_Store"]
```

```python
ocnt = "occurence_count"
tvec = "timing_vectors"
first_user = user_list[0]
path_to_first_user = PurePath(user_json_path, PurePath(f"user_{first_user}.json"))

with open(path_to_first_user, "r") as f:
    user_dict = load(f)
```

```python
word_occurences = [(key, user_dict[key]) for key in user_dict]
word_occurences
```

```python
sorted_word_occurences = sorted(tuple(filter(lambda x: len(x[0]) > 1, [(key, user_dict[key][ocnt]) for key in user_dict])), key = lambda x: x[1], reverse=True)
wordcloud.WordCloud().process_text(user_dict)
```
